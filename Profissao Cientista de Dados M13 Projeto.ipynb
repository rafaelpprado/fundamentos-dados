# 5 - Plote um gráfico de mapa interativo agrupando os dados por categoria, marca e trazendo a média de desconto.

```python
# 5 - Mapa Interativo de Descontos (usando Plotly)

agg_data = df.groupby(['Categoria', 'Marca'])['Desconto'].mean().reset_index()

fig = px.treemap(

  agg_data,

  path=['Categoria', 'Marca'],

  values='Desconto',

  color='Desconto',

  color_continuous_scale='Blues',

  title='Média de Desconto por Categoria e Marca'

)

fig.update_layout(margin=dict(t=50, l=25, r=25, b=25))

fig.show()
```

# 6 - Código todo

```python
import pandas as pd

import matplotlib.pyplot as plt

import plotly.express as px

# Carregar o arquivo CSV

df = pd.read_csv(r'C:\Users\ReReu\PycharmProjects\PythonProject4\modulo7.csv')

# Verificando as colunas do DataFrame

print("Colunas disponíveis no DataFrame:")

print(df.columns)

# Verificando as primeiras linhas do DataFrame

print("\nPrimeiras linhas do DataFrame:")

print(df.head())

# 1 - Média e Mediana dos Preços por Categoria

media_preco = df.groupby('Categoria')['Preco_Normal'].mean().round(2)

mediana_preco = df.groupby('Categoria')['Preco_Normal'].median().round(2)

print("\nMédia dos preços por categoria:\n", media_preco)

print("\nMediana dos preços por categoria:\n", mediana_preco)

# Identificando categorias com média acima ou abaixo da mediana

comparacao = pd.DataFrame({

  'Média': media_preco,

  'Mediana': mediana_preco

})

comparacao['Diferença'] = comparacao['Média'] - comparacao['Mediana']

categorias_especiais = comparacao[comparacao['Diferença'].abs() > 0]

print("\nCategorias com diferença significativa entre média e mediana:")

print(categorias_especiais)

# 2 - Desvio Padrão por Categoria

std_preco = df.groupby('Categoria')['Preco_Normal'].std().round(2)

print("\nDesvio padrão por categoria:\n", std_preco)

# Categoria com maior desvio padrão

cat_maior_std = std_preco.idxmax()

print(f"\nCategoria com maior desvio padrão: {cat_maior_std}")

# Comportamento da média e mediana na categoria com maior desvio

print("\nComportamento na categoria com maior desvio:")

print(comparacao.loc[cat_maior_std])

# 3 - Boxplot da distribuição do Preco_Normal para a categoria com maior desvio padrão

plt.figure(figsize=(10, 6))

plt.boxplot(df[df['Categoria'] == cat_maior_std]['Preco_Normal'],

       patch_artist=True, # preencher o box

       boxprops=dict(facecolor='lightblue'), # cor do box

       medianprops=dict(color='red'), # cor da mediana

       whiskerprops=dict(color='black'), # cor das linhas

       capprops=dict(color='black')) # cor das linhas

plt.title(f'Distribuição de Preços - {cat_maior_std}')

plt.ylabel('Preço Normal')

plt.show()

# 4 - Gráfico de Barras - Média de Descontos por Categoria

media_descontos = df.groupby('Categoria')['Desconto'].mean().round(2)

plt.figure(figsize=(12, 6))

media_descontos.sort_values().plot(kind='barh', color='skyblue')

plt.title('Média de Descontos por Categoria')

plt.xlabel('Média de Desconto (CLP)')

plt.grid(axis='x')

plt.show()

# 5 - Mapa Interativo de Descontos (usando Plotly)

agg_data = df.groupby(['Categoria', 'Marca'])['Desconto'].mean().reset_index()

fig = px.treemap(

  agg_data,

  path=['Categoria', 'Marca'],

  values='Desconto',

  color='Desconto',

  color_continuous_scale='Blues',

  title='Média de Desconto por Categoria e Marca'

)

fig.update_layout(margin=dict(t=50, l=25, r=25, b=25))

fig.show()
```
